# 이항분포 2

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')

#데이터 조건
mu=173
sigma=0.85
x=np.random.normal(mu,sigma,10000)

#히스토그램으로 표현하기
plt.hist(x,bins=30, density=True, edgecolor='k', linewidth=1.2)
plt.title('이항분포')
plt.show()
```

# 정규분포그래프

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rc('font',family='Malgun Gothic')

mu=173
sigma=0.85

x=np.linspace(170,176)
y=1/(sigma * np.sqrt(2 * np.pi))*np.exp(-(x-mu)**2 / (2*sigma**2))
plt.plot(x,y, color='red')
plt.axvline(x=mu, linestyle='--')
plt.xlim(170,176)
plt.show()



```

# 주사위실험

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rc('font', family='HYGothic-Extra')

p=[]
for i in range(2,1001):#주사위를 100번 던지는 시행
    x=list(np.random.randint(1,7,i)) #주사위 1~6까지 중에 랜덤 추출
    x1=x.count(1)# 숫자 1이 나오는 횟수 계산
    p.append(x1/i)# 표본공간의 갯수로 나누어서 확률 계산

plt.scatter(range(2,1001), p,s=0.5)
plt.xlabel('trial')
plt.ylabel('P(x)')
plt.axhline(y=1/6, linestyle='--', color='red')
plt.text(60, 0.25, 'red dot line: 1/6', color='red')
plt.title('주사위를 던져서 1이 나오는 확률')
plt.show()
print(x)
```
