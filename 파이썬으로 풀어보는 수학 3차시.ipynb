```python
import matplotlib.pyplot as plt
t=[53.9,56.3,56.4,53.4,54.5,55.8,56.8,55.0,55.3,54.0,56.7,56.4,57.3]
y=range(2000,2013)
plt.plot(y,t, marker='o', label='F')
plt.legend(loc='best')
```

```python
import matplotlib.pyplot as plt
nyc2000=[31.3,37.3,47.2,51.0,63.5,71.3,72.3,72.7,66.0,57.0,45.3,31.1]
nyc2006=[40.9,35.7,43.1,55.7,63.1,71.0,77.9,75.8,66.6,56.2,51.9,43.6]
month=range(1,13)
plt.plot(month,nyc2000, 'r-.x',linewidth=5, label='2000')
plt.plot(month,nyc2006,'b:o',markersize=10,label='2006')
plt.legend()
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')

nyc2000=[31.3,37.3,47.2,51.0,63.5,71.3,72.3,72.7,66.0,57.0,45.3,31.1]
nyc2006=[40.9,35.7,43.1,55.7,63.1,71.0,77.9,75.8,66.6,56.2,51.9,43.6]
month=range(1,13)
plt.plot(month,nyc2000, marker='o', label='2000')
plt.plot(month,nyc2006, marker='o', label='2006')
plt.legend()
plt.title('뉴욕')
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(6,4),dpi=80)

nyc2000=[31.3,37.3,47.2,51.0,63.5,71.3,72.3,72.7,66.0,57.0,45.3,31.1]
nyc2006=[40.9,35.7,43.1,55.7,63.1,71.0,77.9,75.8,66.6,56.2,51.9,43.6]
month=range(1,13)
plt.plot(month,nyc2000, marker='o', label='2000')
plt.plot(month,nyc2006, marker='o', label='2006')
plt.ylim(0,80)
plt.legend()
plt.title('Newyork City 2000 vs 2006')
```
