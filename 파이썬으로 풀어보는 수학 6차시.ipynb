# 평균값 계산

```python
def calculate_mean(numbers):
    s=sum(numbers)
    n=len(numbers)
    mean=s/n
    return mean
    
donations=[100,60,70,900,100,200,500,500,503,600,1000,1200]

mean=calculate_mean(donations)
print(mean)

```

```python
import numpy as np

donations=[100,60,70,900,100,200,500,500,503,600,1000,1200]
mu=np.mean(donations)
print(mu)
```

# 최빈값을 계산하고 정렬하기

```python
from collections import Counter

list=[4,4,2,2,5,6,1,1,1]

c=Counter(list)
c.most_common()
```

```python
from collections import Counter

score=[7,8,9,2,10,9,9,9,9,4,5,6,1,5,6,7,8,6,1,10]

table=Counter(score)
print('점수\t빈도')
for number in table.most_common():
    print('{0}\t{1}'.format(number[0],number[1]))
```

# 숫자의 분포 및 대푯값

```python
from collections import Counter

score=[7,8,9,2,10,9,9,9,9,4,5,6,1,5,6,7,8,6,1,10]

table=Counter(score)
number_freq=table.most_common()
number_freq.sort()

print('점수\t빈도')
for number in number_freq:
    print('{0}\t{1}'.format(number[0],number[1]))
```

```python
import matplotlib.pyplot as plt
import numpy as np

score=[7,8,9,2,10,9,9,9,9,4,5,6,1,5,6,7,8,6,1,10]

mu=np.mean(score)
sigma=np.std(score)
print('평균은: {0} 표준편차는:{1:.2f}'.format(mu,sigma))

plt.boxplot(score)
plt.grid()
plt.show()
```

# 상관계수를 구하는 수학공식

```python
def find_corr(x,y):
    n=len(x)
    prod=[]
    for xi,yi in zip(x,y):
        prod.append(xi*yi)
        
    sum_prod_xy=sum(prod)
    sum_x=sum(x)
    sum_y=sum(y)
    squared_sum_x=sum_x**2
    squared_sum_y=sum_y**2
    
    x_square=[]
    for xi in x:
        x_square.append(xi**2)
    x_square_sum=sum(x_square)
    
    y_square=[]
    for yi in y:
        y_square.append(yi**2)
    y_square_sum=sum(y_square)
    
    n1=n*sum_prod_xy - sum_x*sum_y
    d1=n*x_square_sum-squared_sum_x
    d2=n*y_square_sum-squared_sum_y
    d=(d1*d2)**0.5
    corr=n1/d
    
    return corr


h=[90,92,95,96,87,87,90,95,98,96]
c=[85,87,86,97,96,88,89,98,98,87]

find_corr(h,c)


    
    
    
    
```
