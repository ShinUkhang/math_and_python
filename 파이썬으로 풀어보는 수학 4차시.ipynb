# 만유인력에 대한 뉴턴의 법칙

```python
import matplotlib.pyplot as plt

r=range(100,1001,50)
g=6.674*(10**-11)
m1=0.5
m2=1.5

f=[]

for d in r:
    force=g*(m1*m2)/(d**2)
    f.append(force)
    
plt.plot(r,f,color='blue', linestyle='--',marker='h',markersize=4)
plt.grid()
plt.title('Gravitational force and diatance')
plt.xlabel('Distance in meters')
plt.ylabel('Gravitational force in newtons')
plt.savefig('만유인력그래프')
plt.show()
```

# 포물선 그리기

```python
import matplotlib.pyplot as plt
import  math
import numpy as np

g=9.8
v=25
theta=45
theta=math.radians(theta)
   
t=2*v*np.sin(theta)/g#체공시간
intervals=np.linspace(0,t,15)
    
x=[]
y=[]
for i in intervals:
    x.append(v*math.cos(theta)*i)
    y.append(v*math.sin(theta)*i- 0.5*g*i*i)
        
plt.plot(x,y,'r--<')
plt.title('Projectile motion of the ball')
plt.grid()
plt.show()
```
