# 대입

```python
from sympy import symbols
x,y=symbols('x,y')
expr=x**2+2*x*y+y**2

res=expr.subs({x:1,y:2})
print(res)
```

```python
from sympy import symbols
x,y=symbols('x,y')
expr=x**2+2*x*y+y**2
expr=expr.subs({x:1-y})
print(expr)
```

# 방정식풀기

```python
from sympy import symbols, solve
x,y=symbols('x,y')
expr=x-5-7
root=solve(expr)
print(root)
```

```python
from sympy import symbols, solve
x,y=symbols('x,y')
expr=x**2+5*x+4 # 3차방정식을 해결해보세요.
root=solve(expr)
print(root)
```

```python
from sympy import symbols, solve
x,y=symbols('x,y')
expr=x**2+5*x+4
root=solve(expr,dict=True)
print(root)
```

```python
from sympy import symbols, solve, pprint
x,y=symbols('x,y')
expr=x**2+x+1
root=solve(expr,dict=True)
print(root)
pprint(root)
```

# 근의 공식

```python
from sympy import symbols, solve,pprint
x,a,b,c=symbols('x,a,b,c')
expr=a*x**2+b*x+c
root=solve(expr,x,dict=True)
print(root)
```

```python
from sympy import symbols, solve,pprint
x,a,b,c=symbols('x,a,b,c')
expr=a*x**2+b*x+c
root=solve(expr,x,dict=True)

pprint(root)
```

# 연립방정식

```python
from sympy import symbols, solve
x,y=symbols('x,y')

expr1=2*x+3*y-6
expr2=3*x+2*y-12

root=solve((expr1,expr2),dict=True)
print(root)
```

# 자연상수e를 표현하기

```python
import math

n=100
x=1
s=0
for i in range(0,n+1):
    s+=x**i/math.factorial(i)
                
print('{0:.5f}'.format(s))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
import math
plt.rc('font', family='Malgun Gothic')

x=np.linspace(100, 1000,1000)
y=[(1+1/i)**i for i in x]

n=sp.Symbol('n')
a=sp.Limit((1+1/n)**n, n, sp.S.Infinity).doit()

plt.plot(x,y)
plt.axhline(y=a, color='red', linestyle='--', label=a)
plt.title('자연상수로의 수렴')
plt.legend()
plt.show()
```
