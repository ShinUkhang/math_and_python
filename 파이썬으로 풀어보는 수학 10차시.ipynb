# 삼각함수

```python
import numpy as np 
import matplotlib.pyplot as plt
from sympy import *    #미적분과 같은 수학계산을 할 때 필요합니다.
plt.rc('font',family='HYGothic-Extra',size=12) #글자체와 크기를 설정합니다. 
plt.rcParams['axes.unicode_minus']=False #음수표현이 가능합니다.

#함수
x=np.linspace(-2*np.pi,2*np.pi,100) # 10개의 구간으로 나누어줍니다.
f=lambda x:sin(x) #람다함수는 수학식을 표현할 때 활용
y=[f(i) for i in x]

plt.scatter(x,y, label='sin',s=2)
plt.grid()
plt.axhline(y=0, color='r', linestyle='--')
plt.axvline(x=0, color='r')
plt.legend()
plt.show()
```

# 미분그래프

```python
import numpy as np 
import matplotlib.pyplot as plt
from sympy import * #미적분과 같은 수학계산을 할 때 필요합니다.
plt.rc('font',family='HYGothic-Extra',size=12) #글자체와 크기를 설정합니다. 
plt.rcParams['axes.unicode_minus']=False #음수표현이 가능합니다.

#함수
x=np.linspace(-3,3,100) #-1부터 3까지 1000개의 구간으로 나누어줍니다.
f=lambda x:x**3 #람다함수는 수학식을 표현할 때 활용
y1=[f(i) for i in x]

t=symbols('t') #수식작업 때 활용할 변수를 이렇게 표현하기도 한다. 
der = diff(t**3,t)
y2=[der.evalf(subs={t:i}) for i in x] #t 대신에 x를 대입해서 미분한다.

plt.plot(x,y1,label='f')
plt.plot(x,y2, label='f_prime')
plt.legend()
plt.show()
```

# 이계도함수

```python
import numpy as np 
import matplotlib.pyplot as plt
from sympy import * #미적분과 같은 수학계산을 할 때 필요합니다.
plt.rc('font',family='HYGothic-Extra',size=12) #글자체와 크기를 설정합니다. 
plt.rcParams['axes.unicode_minus']=False #음수표현이 가능합니다.

#함수
x=np.linspace(-3,3,100) #-1부터 3까지 1000개의 구간으로 나누어줍니다.
f=lambda x:x**3 #람다함수는 수학식을 표현할 때 활용
y1=[f(i) for i in x]

t=symbols('t') #수식작업 때 활용할 변수를 이렇게 표현하기도 한다. 
der = diff(t**3,t)
y2=[der.evalf(subs={t:i}) for i in x] #t 대신에 x를 대입해서 미분한다.

s=symbols('s') #수식작업 때 활용할 변수를 이렇게 표현하기도 한다. 
der = diff(s**3,s,2)
y3=[der.evalf(subs={s:i}) for i in x] #s 대신에 x를 대입해서 미분한다.

plt.plot(x,y1,label='f1')
plt.plot(x,y2, label='f2')
plt.plot(x,y3, label='f3')
plt.axvline(x=0, color='k')
plt.axhline(y=0,color='k')
plt.legend()
plt.show()
```
