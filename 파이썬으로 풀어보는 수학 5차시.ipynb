# 일주일동안 걷는 걸음수

```python
import matplotlib.pyplot as plt

def creat_bar_chart(data,labels):
    num_bars=len(data)
    positions=range(1,num_bars+1)
    plt.barh(positions,data,align='center',color='g', edgecolor='k',alpha=0.1)#가운데로 정렬시킨 역할
    plt.yticks(positions,labels)#y축의 눈금을 표시
    plt.xlabel('Steps')
    plt.ylabel('Day')
    plt.title('Number of steps walked')
    plt.grid()#눈금을 표시하는 역할
    plt.show()
    
data=[6534,7000,8900,10786,3467,11045,5095]
labels=['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
creat_bar_chart(data,labels)
```

# 피보나치수열과 황금비

```python
def fibo(n):
    if n==1:
        return[1]
    if n==2:
        return[1,1]
    a=1
    b=1
    series=[a,b]
    for i in range(n):
        c=a+b
        series.append(c)
        a=b
        b=c
    return series
fibo(10)
```

```python
def fibo(n):
    if n==1:
        return 0
    if n==2:
        return 1
    a=1
    b=1
    ratio=[]
    for i in range(n):
        c=a+b
        ratio.append(b/a)
        a=b
        b=c
    return ratio

import matplotlib.pyplot as plt
plt.plot(fibo(30), 'r--o')
plt.title('Ratio between consecutive Fibonacci numbers')
plt.xlabel('No')
plt.ylabel('Ratio')
plt.grid()
plt.figure(figsize=(10,5),dpi=120)
print('황금비는','{0:.5f}'.format(fibo(30)[29]))
plt.show()
```
